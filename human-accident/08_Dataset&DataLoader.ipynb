ë‹¤ìŒ ë§¤ê°œë³€ìˆ˜ë“¤ì„ ì‚¬ìš©í•˜ì—¬ [FashionMNIST ë°ì´í„°ì…‹](https://pytorch.org/vision/stable/datasets.html#fashion-mnist) ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤:

 - ``download=True`` ëŠ” ``root`` ì— ë°ì´í„°ê°€ ì—†ëŠ” ê²½ìš° ì¸í„°ë„·ì—ì„œ ë‹¤ìš´ë¡œë“œí•©ë‹ˆë‹¤.

    - t10k-labels-idx1-ubyte.gz â†’ í…ŒìŠ¤íŠ¸ìš© ë¼ë²¨

```python
import torch

# PyTorchì—ì„œ ë°ì´í„°ì…‹ì„ í‘œí˜„í•˜ëŠ” ê¸°ë³¸ í´ë˜ìŠ¤ (ì§ì ‘ ìƒì†í•´ì„œ ì»¤ìŠ¤í…€ ê°€ëŠ¥)
from torch.utils.data import Dataset

# torchvisionì´ ì œê³µí•˜ëŠ” ë¯¸ë¦¬ ì¤€ë¹„ëœ ë°ì´í„°ì…‹ ëª¨ìŒ (MNIST, CIFAR10, FashionMNIST ë“±)
from torchvision import datasets

# ì´ë¯¸ì§€ë¥¼ PyTorchì—ì„œ ì“°ê¸° ì¢‹ì€ Tensor íƒ€ì…ìœ¼ë¡œ ë³€í™˜í•´ ì£¼ëŠ” ë„êµ¬
from torchvision.transforms import ToTensor 

import matplotlib.pyplot as plt


training_data = datasets.FashionMNIST(
    root="data",            # root ëŠ” í•™ìŠµ/í…ŒìŠ¤íŠ¸ ë°ì´í„°ê°€ ì €ì¥ë˜ëŠ” ê²½ë¡œ
    train=True,             # train ì€ í•™ìŠµìš© ë˜ëŠ” í…ŒìŠ¤íŠ¸ìš© ë°ì´í„°ì…‹ ì—¬ë¶€ë¥¼ ì§€ì •
    download=True,          # download=True ëŠ” root ì— ë°ì´í„°ê°€ ì—†ëŠ” ê²½ìš° ì¸í„°ë„·ì—ì„œ ë‹¤ìš´ë¡œë“œ
    transform=ToTensor(),   # 
)

test_data = datasets.FashionMNIST(
    root="data",
    train=False,
    download=True,
    transform=ToTensor()
)

# í˜„ì¬ ë””ë ‰í† ë¦¬ ë‚´ data í´ë”ì— ë‹¤ìš´ë¡œë“œ ëœ ê²ƒì„ í™•ì¸
# ê°ê° ë¬´ì—‡ì„ ì˜ë¯¸í•˜ëŠ”ì§€ëŠ” ì˜ ëª¨ë¥´ê² ìŒ.
```

## ğŸŸ¢ ë°ì´í„°ì…‹ì„ ìˆœíšŒí•˜ê³  ì‹œê°í™”í•˜ê¸°  

- ``Dataset`` ì— ë¦¬ìŠ¤íŠ¸(list)ì²˜ëŸ¼ ì§ì ‘ ì ‘ê·¼(index)í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤: ``training_data[index]``.  
- ``matplotlib`` ì„ ì‚¬ìš©í•˜ì—¬ í•™ìŠµ ë°ì´í„°ì˜ ì¼ë¶€ë¥¼ ì‹œê°í™”í•´ë³´ê² ìŠµë‹ˆë‹¤.  



```python
# 0~9 ìˆ«ì ë¼ë²¨ì„ ì‚¬ëŒì´ ì½ì„ ìˆ˜ ìˆëŠ” ë¬¸ìì—´ë¡œ ë°”ê¿”ì£¼ëŠ” ë”•ì…”ë„ˆë¦¬
labels_map = {
    0: "T-Shirt", 
    1: "Trouser", 
    2: "Pullover", 
    3: "Dress", 
    4: "Coat", 
    5: "Sandal", 
    6: "Shirt", 
    7: "Sneaker", 
    8: "Bag", 
    9: "Ankle Boot", 
}  # dictionary

# ì‹œê°í™”í•  ì „ì²´ ê·¸ë¦¼(figure) ê°ì²´ë¥¼ ìƒì„±. figsizeëŠ” ì¸ì¹˜ ë‹¨ìœ„(ê°€ë¡œ,ì„¸ë¡œ).
figure = plt.figure(figsize=(8, 8))  # (8,8)ì€ ì •ì‚¬ê°í˜• ê·¸ë¦¼ì„ ë§Œë“¬.

# 3ì—´ 3í–‰ìœ¼ë¡œ(3x3) ì´ë¯¸ì§€ë¥¼ ë°°ì¹˜í•  ê²ƒì„
cols, rows = 3, 3

# subplot ì¸ë±ìŠ¤ëŠ” 1ë¶€í„° ì‹œì‘( matplotlibì˜ add_subplot ê·œì¹™ )
for i in range(1, cols * rows + 1):
    # training_dataì˜ ì „ì²´ ê¸¸ì´(len)ë¥¼ ë„˜ì§€ ì•ŠëŠ” 0 ì´ìƒ ì •ìˆ˜ ë‚œìˆ˜ í•˜ë‚˜ë¥¼ ë½‘ìŒ.
    # torch.randint(high, size=(1,))ì€ í…ì„œ([idx])ë¥¼ ë°˜í™˜í•˜ë¯€ë¡œ .item()ìœ¼ë¡œ íŒŒì´ì¬ intë¡œ ë³€ê²½
    sample_idx = torch.randint(len(training_data), size=(1,)).item()

    # Datasetì˜ __getitem__ í˜¸ì¶œ: (image, label) íŠœí”Œì„ ë°˜í™˜
    # - image: transform=ToTensor()ê°€ ì ìš©ë˜ì–´ ìˆìœ¼ë¯€ë¡œ shape = [1, H, W] (ì±„ë„, ë†’ì´, ë„ˆë¹„)
    # - label: ì •ìˆ˜(0~9)
    img, label = training_data[sample_idx]

    # figureì— (rows x cols) ê²©ì ì¤‘ ië²ˆì§¸ ìœ„ì¹˜ì— ì„œë¸Œí”Œë¡¯ì„ ì¶”ê°€
    figure.add_subplot(rows, cols, i)

    # í•´ë‹¹ ì´ë¯¸ì§€ì˜ ë¼ë²¨ ë²ˆí˜¸(label)ë¥¼ labels_mapìœ¼ë¡œ ì‚¬ëŒì´ ì½ì„ ìˆ˜ ìˆëŠ” ë¬¸ìì—´ë¡œ ë³€í™˜í•˜ì—¬ ì œëª©ìœ¼ë¡œ í‘œì‹œ
    plt.title(labels_map[label])

    # x, y ì¶• ëˆˆê¸ˆ ë° í…Œë‘ë¦¬ í‘œì‹œë¥¼ ë” (ì´ë¯¸ì§€ ìì²´ë§Œ ë³´ì´ê²Œ)
    plt.axis("off")

    # ì´ë¯¸ì§€ ê·¸ë¦¬ê¸°:
    # - imgëŠ” [1, H, W] ì´ë¯€ë¡œ .squeeze()ë¡œ ì±„ë„ ì°¨ì›(1)ì„ ì œê±° -> [H, W]
    # - cmap="gray"ëŠ” í‘ë°±(ê·¸ë ˆì´ìŠ¤ì¼€ì¼)ë¡œ í‘œì‹œí•˜ê²Œ í•¨
    # ì£¼ì˜: ì¼ë¶€ matplotlib ë²„ì „ì—ì„œëŠ” tensor â†’ numpy ë³€í™˜ì´ í•„ìš”í•  ìˆ˜ ìˆìŒ(img.squeeze().numpy()).
    plt.imshow(img.squeeze(), cmap="gray")

# ëª¨ë“  ì„œë¸Œí”Œë¡¯ì„ í™”ë©´ì— ë³´ì—¬ì¤Œ
plt.show()
```

## ğŸŸ¢ íŒŒì¼ì—ì„œ ì‚¬ìš©ì ì •ì˜ ë°ì´í„°ì…‹ ë§Œë“¤ê¸°  

ì‚¬ìš©ì ì •ì˜ Dataset í´ë˜ìŠ¤ëŠ” ë°˜ë“œì‹œ 3ê°œ í•¨ìˆ˜ë¥¼ êµ¬í˜„í•´ì•¼ í•©ë‹ˆë‹¤: `__init__`, `__len__`, and `__getitem__`.  
ì•„ë˜ êµ¬í˜„ì„ ì‚´í´ë³´ë©´ FashionMNIST ì´ë¯¸ì§€ë“¤ì€ ``img_dir`` ë””ë ‰í† ë¦¬ì— ì €ì¥ë˜ê³ , ì •ë‹µì€ ``annotations_file`` csv íŒŒì¼ì—  
ë³„ë„ë¡œ ì €ì¥ë©ë‹ˆë‹¤.  

ë‹¤ìŒ ì¥ì—ì„œ ê° í•¨ìˆ˜ë“¤ì—ì„œ ì¼ì–´ë‚˜ëŠ” ì¼ë“¤ì„ ìì„¸íˆ ì‚´í´ë³´ê² ìŠµë‹ˆë‹¤.  



```python
import os
import pandas as pd
from torchvision.io import read_image

class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file, names=['file_name', 'label'])
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```

### ``__init__``  

__init__ í•¨ìˆ˜ëŠ” Dataset ê°ì²´ê°€ ìƒì„±(instantiate)ë  ë•Œ í•œ ë²ˆë§Œ ì‹¤í–‰ë©ë‹ˆë‹¤.  
ì—¬ê¸°ì„œëŠ” ì´ë¯¸ì§€ì™€ ì£¼ì„ íŒŒì¼(annotation_file)ì´ í¬í•¨ëœ ë””ë ‰í† ë¦¬ì™€ (ë‹¤ìŒ ì¥ì—ì„œ ìì„¸íˆ ì‚´í´ë³¼) ë‘ê°€ì§€  
ë³€í˜•(transform)ì„ ì´ˆê¸°í™”í•©ë‹ˆë‹¤.  

labels.csv íŒŒì¼ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤: ::  

    tshirt1.jpg, 0  
    tshirt2.jpg, 0  
    ......  
    ankleboot999.jpg, 9  



```python
def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
    self.img_labels = pd.read_csv(annotations_file)
    self.img_dir = img_dir
    self.transform = transform
    self.target_transform = target_transform
```

### ``__len__``  

__len__ í•¨ìˆ˜ëŠ” ë°ì´í„°ì…‹ì˜ ìƒ˜í”Œ ê°œìˆ˜ë¥¼ ë°˜í™˜í•©ë‹ˆë‹¤.  

ì˜ˆ:  



```python
def __len__(self):
    return len(self.img_labels)
```

### ``__getitem__``  

__getitem__ í•¨ìˆ˜ëŠ” ì£¼ì–´ì§„ ì¸ë±ìŠ¤ ``idx`` ì— í•´ë‹¹í•˜ëŠ” ìƒ˜í”Œì„ ë°ì´í„°ì…‹ì—ì„œ ë¶ˆëŸ¬ì˜¤ê³  ë°˜í™˜í•©ë‹ˆë‹¤.  
ì¸ë±ìŠ¤ë¥¼ ê¸°ë°˜ìœ¼ë¡œ, ë””ìŠ¤í¬ì—ì„œ ì´ë¯¸ì§€ì˜ ìœ„ì¹˜ë¥¼ ì‹ë³„í•˜ê³ , ``read_image`` ë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ë¯¸ì§€ë¥¼ í…ì„œë¡œ ë³€í™˜í•˜ê³ , ``self.img_labels`` ì˜ csv ë°ì´í„°ë¡œë¶€í„°  
í•´ë‹¹í•˜ëŠ” ì •ë‹µ(label)ì„ ê°€ì ¸ì˜¤ê³ , (í•´ë‹¹í•˜ëŠ” ê²½ìš°) ë³€í˜•(transform) í•¨ìˆ˜ë“¤ì„ í˜¸ì¶œí•œ ë’¤, í…ì„œ ì´ë¯¸ì§€ì™€ ë¼ë²¨ì„ Python ì‚¬ì „(dict)í˜•ìœ¼ë¡œ ë°˜í™˜í•©ë‹ˆë‹¤.  



```python
def __getitem__(self, idx):
    img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
    image = read_image(img_path)
    label = self.img_labels.iloc[idx, 1]
    if self.transform:
        image = self.transform(image)
    if self.target_transform:
        label = self.target_transform(label)
    sample = {"image": image, "label": label}
    return sample
```

## ğŸŸ¢ DataLoaderë¡œ í•™ìŠµìš© ë°ì´í„° ì¤€ë¹„í•˜ê¸°  

``Dataset`` ì€ ë°ì´í„°ì…‹ì˜ íŠ¹ì§•(feature)ì„ ê°€ì ¸ì˜¤ê³  í•˜ë‚˜ì˜ ìƒ˜í”Œì— ì •ë‹µ(label)ì„ ì§€ì •í•˜ëŠ” ì¼ì„ í•œ ë²ˆì— í•©ë‹ˆë‹¤.  
ëª¨ë¸ì„ í•™ìŠµí•  ë•Œ, ì¼ë°˜ì ìœ¼ë¡œ ìƒ˜í”Œë“¤ì„ "ë¯¸ë‹ˆë°°ì¹˜(minibatch)"ë¡œ ì „ë‹¬í•˜ê³ , ë§¤ ì—í­(epoch)ë§ˆë‹¤ ë°ì´í„°ë¥¼ ë‹¤ì‹œ ì„ì–´ì„œ ê³¼ì í•©(overfit)ì„ ë§‰ê³ ,  
Pythonì˜ ``multiprocessing`` ì„ ì‚¬ìš©í•˜ì—¬ ë°ì´í„° ê²€ìƒ‰ ì†ë„ë¥¼ ë†’ì´ë ¤ê³  í•©ë‹ˆë‹¤.  

``DataLoader`` ëŠ” ê°„ë‹¨í•œ APIë¡œ ì´ëŸ¬í•œ ë³µì¡í•œ ê³¼ì •ë“¤ì„ ì¶”ìƒí™”í•œ ìˆœíšŒ ê°€ëŠ¥í•œ ê°ì²´(iterable)ì…ë‹ˆë‹¤.  



```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(training_data, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_data, batch_size=64, shuffle=True)
```

## ğŸŸ¢ DataLoaderë¥¼ í†µí•´ ìˆœíšŒí•˜ê¸°(iterate)  

``DataLoader`` ì— ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜¨ ë’¤ì—ëŠ” í•„ìš”ì— ë”°ë¼ ë°ì´í„°ì…‹ì„ ìˆœíšŒ(iterate)í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.  
ì•„ë˜ì˜ ê° ìˆœíšŒ(iteration)ëŠ” (ê°ê° ``batch_size=64`` ì˜ íŠ¹ì§•(feature)ê³¼ ì •ë‹µ(label)ì„ í¬í•¨í•˜ëŠ”) ``train_features`` ì™€  
``train_labels`` ì˜ ë¬¶ìŒ(batch)ì„ ë°˜í™˜í•©ë‹ˆë‹¤. ``shuffle=True`` ë¡œ ì§€ì •í–ˆìœ¼ë¯€ë¡œ, ëª¨ë“  ë°°ì¹˜ë¥¼ ìˆœíšŒí•œ ë’¤ ë°ì´í„°ê°€ ì„ì…ë‹ˆë‹¤.  
(ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° ìˆœì„œë¥¼ ë³´ë‹¤ ì„¸ë°€í•˜ê²Œ(finer-grained) ì œì–´í•˜ë ¤ë©´ [Samplers](https://pytorch.org/docs/stable/data.html#data-loading-order-and-sampler)  
ë¥¼ ì‚´í´ë³´ì„¸ìš”.)  



```python
# ì´ë¯¸ì§€ì™€ ì •ë‹µ(label)ì„ í‘œì‹œí•©ë‹ˆë‹¤.
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")

img = train_features[0].squeeze()
label = train_labels[0]

plt.imshow(img, cmap="gray")
plt.show()
print(f"Label: {label}")
```
