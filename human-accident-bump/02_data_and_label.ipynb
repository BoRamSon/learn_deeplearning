### üü° 4-1. OpenCV Í∑∏Î¶¨Í∏∞ Ìï®Ïàò  

- OpenCVÎäî ÏòÅÏÉÅÏóêÏÑú ÏÑ†, ÎèÑÌòï, Î¨∏ÏûêÏó¥ÏùÑ Ï∂úÎ†•ÌïòÎäî Í∑∏Î¶¨Í∏∞ Ìï®ÏàòÎ•º Ï†úÍ≥µ  
- ÏÑ† Í∑∏Î¶¨Í∏∞(ÏßÅÏÑ†, ÌôîÏÇ¥Ìëú, ÎßàÏª§), ÎèÑÌòï Í∑∏Î¶¨Í∏∞(ÏÇ¨Í∞ÅÌòï, Ïõê, ÌÉÄÏõê, Îã§Í∞ÅÌòï), Î¨∏ÏûêÏó¥ Ï∂úÎ†• Ìï®ÏàòÎ•º ÏïåÏïÑÎ≥¥Í≥† ÏòÅÏÉÅÏóê Ï†ÅÏö©Ìï¥Î≥¥Ïûê!  

#### ÌïòÏßÄÎßå Îã§ Í±¥ÎÑàÎõ∞Í≥† ÏùºÎã® **`Ï†êÏ∞çÍ∏∞`** Î∂ÄÌÑ∞ Ìï¥Î≥¥Ïûê  
- circleÏù¥ÏßÄÎßå Ïù¥Í±∏ Ï†êÏúºÎ°ú ÏÉùÍ∞ÅÌïòÍ≥† Ï∞çÎäîÎã§.  
- cv2.circle(img, center, radius, color, thickness=None, lineType=None, shift=None) -> img  


```python
# imreadÎ°ú Ïù¥ÎØ∏ÏßÄÎ•º Í∞ÄÏ†∏ÏôÄÏÑú imageÎùºÎäî Î≥ÄÏàòÏóê Ï†ÄÏû•ÌïòÏòÄÏóàÎã§.
drawing_image = image.copy()  # ÏõêÎ≥∏ÏùÑ Ìï¥ÏπòÏßÄ ÏïäÍ∏∞ ÏúÑÌï¥ Ìï¥Îãπ Ïù¥ÎØ∏ÏßÄÎ•º copy Ìï¥ÏôîÏùå.

# üî• Î®ºÏ†Ä ÌïòÎìúÏΩîÎî©ÏùÑ ÌÜµÌï¥ÏÑú Ï∂úÎ†• Ïó¨Î∂ÄÎßå Î®ºÏ†Ä ÌôïÏù∏ÏùÑ Ìï¥Î≥¥Í≤†ÏäµÎãàÎã§.
# cv2.circle(img, center(x,yÍ∞í), radius(ÏõêÏùò Î∞òÏßÄÎ¶Ñ), color(R,G,B),
#               thickness=None(ÎëêÍªò), lineType=None, shift=None) -> img
# cv2.circle(drawing_image, (580, 600), 200, (255, 0, 0), 30)
cv2.circle(drawing_image, (580, 600), 8, (255, 0, 0), -1)
cv2.circle(drawing_image, (550, 600), 8, (255, 0, 0), -1)
cv2.circle(drawing_image, (530, 600), 8, (255, 0, 0), -1)

# cv2.imshow("image_window", drawing_image)  # ÏùëÎãµÏóÜÏùåÏúºÎ°ú Ïù∏Ìï¥ matplotlib ÏÇ¨Ïö©Ìï®.
plt.imshow(drawing_image)  # Ïûò Í∑∏Î†§ÏßÄÎäî Í≤ÉÏùÑ ÌôïÏù∏ÌñàÏäµÎãàÎã§.
```

### üü° 4-2. Î≥∏Í≤©Ï†ÅÏúºÎ°ú segmentation Í∞íÏùÑ Í∞ÄÏßÄÍ≥† Ï†ê Ï∞çÏñ¥Î≥¥Í∏∞  

```python
# imreadÎ°ú Ïù¥ÎØ∏ÏßÄÎ•º Í∞ÄÏ†∏ÏôÄÏÑú imageÎùºÎäî Î≥ÄÏàòÏóê Ï†ÄÏû•ÌïòÏòÄÏóàÎã§.
segmentation_image = image.copy()  # ÏõêÎ≥∏ÏùÑ copy Ìï¥Ïò¥.

# cv2.circle(img, center(x,yÍ∞í), radius(ÏõêÏùò Î∞òÏßÄÎ¶Ñ), color(R,G,B),
#               thickness=None(ÎëêÍªò), lineType=None, shift=None) -> img
for i in range(len(x)):
    # üî•floatÏù¥Î©¥ overload ÏóêÎü¨ Î∞úÏÉù
    # > Overload resolution failed:
    # >  - Can't parse 'center'. Sequence item with index 0 has a wrong type
    # >  - Can't parse 'center'. Sequence item with index 0 has a wrong type
    # = Í≥µÏãù documentsÎ•º Î≥¥Î©¥ intÎ°úÎßå Î∞õÎäî Í≤ÉÏúºÎ°ú ÎêòÏñ¥ÏûàÍ∏∞ ÎïåÎ¨∏Ïóê intÎ°ú Î∞îÍøîÏ£ºÎäî Í≤ÉÏù¥ ÎßûÏäµÎãàÎã§.
    cv2.circle(segmentation_image, (int(x[i]), int(y[i])), 8, (255, 0, 0), -1)

plt.imshow(segmentation_image)
```

### üü° 5-1. OpenCV - rectangle() Í∑∏Î†§Î≥¥Í∏∞  

- cv2.rectangle(img, pt1, pt2, color, thickness=None, lineType=None, shift=None) -> img  
- cv2.rectangle(img, rec, color, thickness=None, lineType=None, shift=None) -> img  
    - img : Í∑∏Î¶ºÏùÑ Í∑∏Î¶¥ ÏòÅÏÉÅ  
    - pt1, pt2 : ÏÇ¨Í∞ÅÌòïÏùò Îëê Íº≠ÏßÄÏ†ê Ï¢åÌëú. (x, y) ÌäúÌîå  
    - rec : ÏÇ¨Í∞ÅÌòï ÏúÑÏπò Ï†ïÎ≥¥. (x, y, w, h) ÌäúÌîå  
    - color : ÏÑ† ÏÉâÏÉÅ ÎòêÎäî Î∞ùÍ∏∞. (B, G, R) ÌäúÌîå ÎòêÎäî Ï†ïÏàòÍ∞í  
    - thickness : ÏÑ† ÎëêÍªò. Í∏∞Î≥∏Í∞íÏùÄ 1. ÏùåÏàò(-1)Î•º ÏßÄÏ†ïÌïòÎ©¥ ÎÇ¥Î∂ÄÎ•º Ï±ÑÏõÄ.  
    - shift : Í∑∏Î¶¨Í∏∞ Ï¢åÌëú Í∞íÏùò Ï∂ïÏÜå ÎπÑÏú®. Í∏∞Î≥∏Í∞íÏùÄ 0  



```python
rectangle_image = segmentation_image.copy()
# segmentationÏù¥ ÏôÑÎ£åÎêú Ïù¥ÎØ∏ÏßÄÎ•º Î≥µÏÇ¨Ìï¥Ïò®Îã§.

x_min = min(x)
# print(f"x_min : {x_min}")
x_max = max(x)
# print(f"x_max : {x_max}")
y_min = min(y)
# print(f"y_min : {y_min}")
y_max = max(y)
# print(f"xy_max : {y_max}")


pt1 = (int(x_min), int(y_max))
pt2 = (int(x_max), int(y_min))
cv2.rectangle(rectangle_image, pt1, pt2, (255, 0, 0), 5)

plt.imshow(rectangle_image)
```
